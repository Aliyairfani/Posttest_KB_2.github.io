Untuk menampilkan dataset dengan pandas python, dataset kita simpan pada 1 folder, sebelumnya kita hurus import pandas terlebih dahulu. Lalu kita baca file csv dengan menggunakan sintaks dibawah;


```python
import pandas as pd

bankMarketing = pd.read_csv('bank-additional-full.csv',sep=";")
```

```python
record, attribute = bankMarketing.shape
jml_object = bankMarketing.select_dtypes(["float", "int"]).any().sum()
nama_attribute = bankMarketing.columns.values
jml_null = bankMarketing.isnull().any().sum()
```

```python
info = {"Jumlah Record": record, "Jumlah Attribute" : attribute, "Nama Attribute" : list(nama_attribute), "Jumlah Object" : jml_object, "Jumlah Nilai Null": jml_null}
```

Dibawah ini adalah tampilan dictionary;

```python
for x in info :
    print(x, ":", info[x])
```

Berikut adalah tampilan untuk menunjukan jumlah, rata-rata, simpangan baku, nilai minimal, nilai kuartil dan nilai maksimal;

```python
bankMarketing = bankMarketing.select_dtypes(["float","int"])
bankMarketing.describe()
```

Tampilan dibawah ini merupakan tampilan nilai korelasi attriute bertipe data angka;

```python
bankMarketing.corr()
```

Dibawah ini merupakan sintax untuk mengubah nilai age menjadi menjadi bentuk desimal dangan 3 angka dibelakang koma;

```python
bankMarketing.age = bankMarketing.age.astype("float").apply(lambda x: format(x, ".3f"))
bankMarketing.head(10)
```
