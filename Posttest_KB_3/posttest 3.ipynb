Sebelum melakukan pendeklarasian dengan grafik kita harus mengimport package dibawah ini 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
bankMarketing = pd.read_csv('bank-additional-full.csv',sep=";") #pendeklarasian untuk membaca data
```

```python
bankMarketing.head()
```

```python
bankMarketing.info() 
```

Grafik batang ini menunjukkan umur pelanggan dan status pekerjaan pelanggan. 

```python
x = list (bankMarketing["age"])
y = list (bankMarketing["job"])

plt.figure(figsize =(7, 10))
plt.bar(x, y, color='blue')
plt.xlabel("AGE")
plt.ylabel("JOB")
plt.title("umur dan pekerjaan pelanggan")
plt.show()
```

Grafik pie dibawah ini menunjukkan bahwa pelanggan yang melakukan peminjaman di bank ada sebanyak 15.2%,
yang tidak melakukan peminjaman ada sebanyak 82.4% dan sisanya yaitu 2.4% merupakan nilai null pada data tersebut.

```python
data_pie = bankMarketing['loan'].value_counts().rename('loan').reset_index(name='id_count') 
```

```python
data_pie
```

```python
plt.pie(data_pie.id_count, labels=data_pie["index"], startangle=90,
autopct='%.1f%%', shadow= True, explode= (0.2, 0.1 , 0))
plt.title('pinjaman')
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize =(5, 5))
sns.countplot(data = bankMarketing, x = "marital", hue = "education", palette = "flare")
plt.title('status perkawinan dan akademik')
```

Grafik tersebut menunjukkan status perkawinan dan akademik pelanggan. Ada yang sudah menikah dan memiliki gelar sarjana dan lainnya, ada juga yang single dan memiliki status akademik yang baik dan ada juga yang telah cerai dengan status akademik yang terbilang masih dasar

```python
sns.boxplot(data = bankMarketing, x = 'housing', y = 'campaign')
```

Grafik  diatas menunjukkan data perumahan dan data periklanan, ada beberapa yang memiliki dan ada beberapa yang tidak memiliki perumahan dan periklanan untuk mengetahui bank ini, dan ada yang tdak tau
