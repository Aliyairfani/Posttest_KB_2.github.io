## DATA CLEANING

### NILAI NULL

```python
startup_data2 = pd.read_csv('startup data rusak.csv', sep=";")
```

```python
data_bersih = startup_data2
```

```python
data_bersih.isnull().sum()
```

```python
data_bersih[0:61]
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
```

```python
data_bersih["has_roundA"] = imputer.fit_transform(data_bersih[["has_roundA"]])
```

```python
data_bersih.isnull().sum()
```

```python
data_bersih["has_roundB"]=data_bersih["has_roundB"].fillna(data_bersih["has_roundB"].mean())
```

```python
data_bersih.isnull().sum()
```

```python
data_bersih.fillna(data_bersih.mean())
data_bersih.isna().sum()
```

### NILAI DUPLIKAT

```python
data_bersih[data_bersih.duplicated()]
```

```python
data_bersih.drop_duplicates(inplace=True)
data_bersih.duplicated().sum()
```
